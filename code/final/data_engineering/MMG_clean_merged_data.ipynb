## Clean County Data

```python
county_df = pd.read_csv("MMG_county_data_all_years.csv")
county_df.head()
```

```python
# Fixing rows with no state abbreviation
def fix_state_abbreviation(row):
    if row['State_Abbreviation'] is np.nan:
        return row['State Name']
    else:
        return row['State_Abbreviation']
county_df['State_Abbreviation'] = county_df.apply(fix_state_abbreviation, axis = 1)
del county_df['State Name']

# Discovered that 68 rows have mostly NaN values. We'll just remove them
county_df = county_df.dropna(subset=['State_Abbreviation'])

county_df.head()
```

```python
# There's 2 different columns for country. Let's make it just one
def fix_county_code(row):
    if row['County_State_Name'] is np.nan and row['County_Code'] is np.nan:
        return np.nan
    elif row['County_State_Name'] is np.nan:
        return row['County_Code'].lower()
    else:
        return " ".join(row['County_State_Name'].split(",")[0].split(" ")[:-1]).lower()
county_df['County_Code'] = county_df.apply(fix_county_code, axis = 1)
del county_df['County_State_Name']

county_df
```

```python
len(county_df[county_df['Annual_Food_Budget_Shortfall'].isnull()])
```

```python
def fix_list_cols(row):
    var_name = 'Low_Threshold_Type'
    if len(row[var_name].split(", ")) > 1:
        return row[var_name].split(", ")[0]
    else:
        return row[var_name]
county_df['Low_Threshold_Type'] = county_df.apply(fix_list_cols, axis = 1)

def fix_list_cols(row):
    var_name = 'High_Threshold_Type'
    if len(row[var_name].split(", ")) > 1:
        return row[var_name].split(", ")[0]
    else:
        return row[var_name]
county_df['High_Threshold_Type'] = county_df.apply(fix_list_cols, axis = 1)

county_df.head()
```

```python
identifying_cols = [
    'Year',
    'FIPS',
    'County_Code',
    'State_Abbreviation'
]
other_cols = [col for col in county_df.columns if col not in identifying_cols]
new_column_order = identifying_cols + other_cols
county_df = county_df[new_column_order]
county_df.rename({'County_Code': 'County', 'State_Abbreviation': 'State'}, axis = 1, inplace = True)
county_df.head()
```

```python
print("Null Value Counts For County DF:\n")
for col in county_df.columns:
    print(col, len(county_df[county_df[col].isnull()]))
```

```python
county_df.to_csv("MMG_county_df_clean.csv", index = False)
```

## Clean District Data

```python

district_df = pd.read_csv("MMG_district_data_all_years.csv")
district_df.head()
```

```python
print("Null Value Counts:\n")
for col in district_df.columns:
    print(col, len(district_df[district_df[col].isnull()]))
```

```python
def fix_list_cols(row):
    var_name = 'Low_Threshold_Type'
    if len(row[var_name].split(", ")) > 1:
        return row[var_name].split(", ")[0]
    else:
        return row[var_name]
district_df['Low_Threshold_Type'] = district_df.apply(fix_list_cols, axis = 1)

def fix_list_cols(row):
    var_name = 'High_Threshold_Type'
    if len(row[var_name].split(", ")) > 1:
        return row[var_name].split(", ")[0]
    else:
        return row[var_name]
district_df['High_Threshold_Type'] = district_df.apply(fix_list_cols, axis = 1)
```

```python
print("Mean Below Th: ", district_df['Pct_FI_Below_Low_Threshold'].mean())
print("Mean SNAP_Eligable: ", district_df['SNAP_Eligible'].mean())
print()
print("Mean Between Th: ", district_df['Pct_FI_Between_Thresholds'].mean())
print("Mean Only_NSLP_WIC_Eligible: ", district_df['Only_NSLP_WIC_Eligible'].mean())
print()
print("Mean Above Th: ", district_df['Pct_FI_Above_High_Threshold'].mean())
print("Mean No_Eligibility: ", district_df['No_Eligibility'].mean())
```

```python
def fix_blw_th(row):
    if np.isnan(row['Pct_FI_Below_Low_Threshold']):
        return row['SNAP_Eligible']
    else:
        return row['Pct_FI_Below_Low_Threshold']
district_df['Pct_FI_Below_Low_Threshold'] = district_df.apply(fix_blw_th, axis = 1)

def fix_btw_th(row):
    if np.isnan(row['Pct_FI_Between_Thresholds']):
        return row['Only_NSLP_WIC_Eligible']
    else:
        return row['Pct_FI_Between_Thresholds']
district_df['Pct_FI_Between_Thresholds'] = district_df.apply(fix_btw_th, axis = 1)

def fix_abv_th(row):
    if np.isnan(row['Pct_FI_Above_High_Threshold']):
        return row['No_Eligibility']
    else:
        return row['Pct_FI_Above_High_Threshold']
district_df['Pct_FI_Above_High_Threshold'] = district_df.apply(fix_abv_th, axis = 1)

del district_df['SNAP_Eligible']
del district_df['Only_NSLP_WIC_Eligible']
del district_df['No_Eligibility']

district_df.head()
```

```python
print("Null Value Counts For District DF:\n")
for col in district_df.columns:
    print(col, len(district_df[district_df[col].isnull()]))
```

```python
district_df.to_csv("MMG_district_df_clean.csv", index = False)
```
