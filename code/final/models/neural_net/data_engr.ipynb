```python
import pandas as pd
import numpy as np
import os
os.chdir('/home/blalo/uva/ml1/ml_group_project_fall_2025/code/final')
df = pd.read_csv('data/data_w_clusters.csv')
df.head().T
```

```python
def add_last_year_cluster(df):

    """

    Creates a 'last_year_cluster' feature by matching FIPS and YEAR-1.

    """

    previous_year_data = df[['FIPS', 'YEAR', 'cluster']].copy()

    previous_year_data['YEAR'] += 1

    previous_year_data = previous_year_data.rename(columns={'cluster': 'last_year_cluster'})

    df_merged = pd.merge(

        df,

        previous_year_data,

        on=['FIPS', 'YEAR'],

        how='left'

    )



    return df_merged



df_result = add_last_year_cluster(df)

df_result.sort_values(by=['FIPS', 'YEAR']).T
```

```python
df_result = df_result.dropna()
df_result.to_csv('data/nn_df.csv', index=False)
```
